```python
# Verificar a quantidade de valores nulos (NA)
df.isnull().sum()
```

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Verificar a existência de outliers (usando o método de quantis)
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1
outliers = ((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR))).sum()
print("Outliers por coluna:\n", outliers)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Convertendo colunas numéricas que podem ter sido carregadas como string
df['TOTAL_VACCINATIONS'] = pd.to_numeric(df['TOTAL_VACCINATIONS'], errors='coerce')
df['PERSONS_VACCINATED_1PLUS_DOSE'] = pd.to_numeric(df['PERSONS_VACCINATED_1PLUS_DOSE'], errors='coerce')
df['PERSONS_LAST_DOSE'] = pd.to_numeric(df['PERSONS_LAST_DOSE'], errors='coerce')
df['PERSONS_BOOSTER_ADD_DOSE'] = pd.to_numeric(df['PERSONS_BOOSTER_ADD_DOSE'], errors='coerce')
```

```python
from pickle import FALSE
```

```python
# Gráfico de Barras: Total de pessoas vacinadas por país
plt.figure(figsize=(12, 8))
sns.barplot(x='COUNTRY', y='PERSONS_VACCINATED_1PLUS_DOSE', data=df.sort_values(by=['PERSONS_VACCINATED_1PLUS_DOSE'],ascending=False)[:50])
plt.xticks(rotation=90)
plt.title("Total de Pessoas Vacinas com pelo menos 1 dose por País")
plt.show()
```

```python

# Linha do Tempo: Progresso da vacinação ao longo do tempo
df['FIRST_VACCINE_DATE'] = pd.to_datetime(df['FIRST_VACCINE_DATE'], errors='coerce')

plt.figure(figsize=(12, 6))
df.groupby('FIRST_VACCINE_DATE')['PERSONS_VACCINATED_1PLUS_DOSE'].sum().plot()
plt.title('Progresso da Vacinação ao Longo do Tempo')
plt.xlabel('Data da Primeira Vacinação')
plt.ylabel('Pessoas Vacinadas (1+ Dose)')
plt.show()
```

```python
# Proporção de pessoas que completaram a vacinação (última dose)
df['FULLY_VACCINATED_PERCENTAGE'] = (df['PERSONS_LAST_DOSE'] / df['PERSONS_VACCINATED_1PLUS_DOSE']) * 100

# Histograma para a nova métrica
plt.figure(figsize=(10, 6))
sns.histplot(df['FULLY_VACCINATED_PERCENTAGE'].dropna(), bins=20, kde=True)
plt.title('Distribuição da Percentagem de Pessoas Completamente Vacinadas')
plt.xlabel('Percentual de Pessoas Completamente Vacinadas')
plt.ylabel('Frequência')
plt.show()
```

```python
# Remover valores nulos da coluna PERSONS_LAST_DOSE_PER100
df_clean = df.dropna(subset=['PERSONS_LAST_DOSE_PER100'])
```

```python
# Definir um limite superior para evitar outliers extremos no gráfico (ex: 100%)
df_clean = df_clean[df_clean['PERSONS_LAST_DOSE_PER100'] <= 100]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot: Distribuição da porcentagem de pessoas que receberam a última dose
plt.figure(figsize=(10, 6))
sns.boxplot(x='WHO_REGION', y='PERSONS_LAST_DOSE_PER100', data=df_clean)
plt.title('Distribuição de Pessoas com Última Dose por Região da OMS (Limite de 100%)')
plt.xlabel('Região OMS')
plt.ylabel('Percentual de Pessoas com Última Dose')
plt.show()
```

```python
import numpy as np
# Aplicar transformação logarítmica na coluna PERSONS_LAST_DOSE_PER100
df_clean['LOG_PERSONS_LAST_DOSE_PER100'] = df_clean['PERSONS_LAST_DOSE_PER100'].apply(lambda x: np.log1p(x))

# Boxplot com transformação logarítmica
plt.figure(figsize=(10, 6))
sns.boxplot(x='WHO_REGION', y='LOG_PERSONS_LAST_DOSE_PER100', data=df_clean)
plt.title('Distribuição Logarítmica de Pessoas com Última Dose por Região da OMS')
plt.xlabel('Região OMS')
plt.ylabel('Log(Percentual de Pessoas com Última Dose)')
plt.show()
```
